```python
import pandas as pd
import matplotlib.pyplot as plt

x = pd.read_excel(r'..\csv\datas_name_age_and_region.xlsx')

print(x)

plt.style.use('seaborn-darkgrid') # Define o fundo do gráfico
plt.figure(figsize=(8,5)) # Define o tamanho do gráfico

plt.hist(x['Idade'], bins=20, color='green') # Estipula as barras com um certo distanciamento
plt.xlabel('Valores de idade', fontsize=13, fontfamily='serif')
plt.ylabel('Pessoas com mesma idade', fontsize=13, fontfamily='serif')

plt.show()
```

```python
plt.figure(figsize=(8,5)) # Define o tamanho do gráfico

plt.pie(x['Idade'], labels= x['Nome'], autopct='%1.2f%%') # Estipula a idade em porcentagem

plt.show()
```

```python
plt.style.use('seaborn-darkgrid') # Define o fundo do gráfico
plt.figure(figsize=(8,5)) # Define o tamanho do gráfico

plt.xlabel('Valores de Altura', fontsize=13, fontfamily='serif')
plt.ylabel('Valores de Idade', fontsize=13, fontfamily='serif')

age = list(str(x.Idade).split('\n'))
age.pop(-1)

height = list(str(x.Altura).split('\n'))

name = list(str(x.Nome).split('\n'))


color = ['orange', 'black', 'blue', 'darkslategrey', 'red', 'purple', 'lime']

lista_tanto = []
for i in age:
    a, b = i.split('    ')
    lista_tanto.append(int(b))
    
lista_tanto = sorted(lista_tanto)
print(lista_tanto)


for value in range(7):
    plt.barh(str(age[0 + value][1:]),
             height[0 + value][1:],
             color=color[0 + value],
             label=name[0 + value][4:])

    plt.legend(fontsize=13, frameon=True, framealpha=0.2, facecolor='grey')

plt.show()
```
