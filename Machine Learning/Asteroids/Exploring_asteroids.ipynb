# Explorando NASA JPL Asteroid Dataset 

Dataset disponivel em Kaggle: https://www.kaggle.com/sakhawat18/asteroid-dataset

### Resumo
 
- SPK-ID: Object primary SPK-ID
- Object ID: Object internal database ID
- Object fullname: Object full name/designation
- pdes: Object primary designation
- name: Object IAU name
- NEO: Near-Earth Object (NEO) flag
- PHA: Potentially Hazardous Asteroid (PHA) flag
- H: Absolute magnitude parameter
- Diameter: object diameter (from equivalent sphere) km Unit
- Albedo: Geometric albedo
- Diameter_sigma: 1-sigma uncertainty in object diameter km Unit
- Orbit_id: Orbit solution ID
- Epoch: Epoch of osculation in modified Julian day form
- Equinox: Equinox of reference frame
- e: Eccentricity
- a: Semi-major axis au Unit
- q: perihelion distance au Unit
- i: inclination; angle with respect to x-y ecliptic plane
- tp: Time of perihelion passage TDB Unit
- moid_ld: Earth Minimum Orbit Intersection Distance au Unit

```python
# Nulos em colunas sigmas, semelhante a coluna pha e moid
asteroids1[[ 'sigma_e','sigma_a','sigma_q','sigma_i','sigma_om',
             'sigma_w','sigma_ma','sigma_ad','sigma_n','sigma_tp','sigma_per' ]].isnull().sum()
```

```python
# Porcentagem de nulos em colunas sigmas, semelhante a 2% de pha e moid
round((asteroids1[[ 'sigma_e','sigma_a','sigma_q',
                    'sigma_i','sigma_om','sigma_w',
                    'sigma_ma','sigma_ad','sigma_n',
                    'sigma_tp','sigma_per' ]].isnull().sum() / asteroids1['pha'].shape[0])*100, 2)
```

```python
# porcentagem de nulos nas colunas 
round((asteroids1[['diameter', 'albedo', 'diameter_sigma']].isnull().sum() / asteroids1['pha'].shape[0])*100, 2)
```

```python
# Removendo Colunas Diameter, Albedo e Diameter_sigma contem muitos valores nulos, 85% em media são nulos
asteroids1 = asteroids1.drop(columns=['diameter', 'albedo', 'diameter_sigma'], axis=1)

# Retornando linhas com valores, assim removendo da busca os 2% nulos gerais de pha, moid e sigmas
asteroids2 = asteroids1[asteroids1['pha'].notna()]
asteroids2.info()
```

```python
# Na coluna 'H' - magnitude absoluta, contem alguns nulos. 
# Ela pode ser obtida atraves do albedo e diâmetro, mas as colunas foram removidas, 
# então, podemos remover as linhas nulas.
asteroids2 = asteroids2[asteroids2['H'].notna()]
asteroids2.info()
```

```python
# Para as colunas restantes, a referencia de sigma_ad contem o menor numeros de dados
# assim, retornamos a base de dados referente a esta coluna mais limpa
asteroids2 = asteroids2[asteroids2['sigma_ad'].notna()]
asteroids2.info()
```

```python
# Visualizando os retornos, existe uma coluna 'ma' com 1 dado faltantes, tambem removeremos ele.
asteroids2 = asteroids2[asteroids2['ma'].notna()]
asteroids2.info()
```

```python
asteroids2.head()
```

```python
plt.figure(figsize=(30,25));

sns.heatmap(asteroids2.corr(), annot=True);
```

```python
# pha = Asteroides Potencialmente Perigosos, Yes ou No

plt.figure(figsize=(15,10))
plot = sns.countplot(x='pha', data=asteroids2);

for i in plot.patches:
    plot.annotate(format(i.get_height(), '.2f'), (i.get_x() + i.get_width() / 2., i.get_height()),
                   ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.xlabel('PHA - Asteroides Potencialmente Perigosos', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.title('Quantidade de Asteroides Perigosos', fontsize=16)
plt.show()
```

```python
# neo = Proximos da Terra, Yes ou No

plt.figure(figsize=(15,10))
plot = sns.countplot(x='neo', data=asteroids2);

for i in plot.patches:
    plot.annotate(format(i.get_height(), '.2f'), (i.get_x() + i.get_width() / 2., i.get_height()),
                   ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.xlabel('Asteroides Proximos da Terra', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.title('Quantidade de Asteroides Proximos da Terra', fontsize=16)
plt.show()
```

```python
# pha = Asteroides Potencialmente Perigosos, Yes ou No
# neo = Proximos da Terra, Yes ou No

plt.figure(figsize=(12,10))
plot = sns.countplot(x='pha', hue='neo', data=asteroids2);

for i in plot.patches:
    plot.annotate(format(i.get_height(), '.2f'), (i.get_x() + i.get_width() / 2., i.get_height()),
                   ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
    
plt.legend(labels=['Longe Terra', 'Perto Terra'])
plt.xlabel('PHA - Asteroides Potencialmente Perigosos', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.title('Asteroides Perigosos e proximos da Terra', fontsize=16)
plt.show()
```

```python
# neo = Proximos da Terra, Yes ou No
# H = Magnitude Absoluta, medida de brilho do objeto celeste 

plt.figure(figsize=(12,10))

sns.boxplot(x="neo", y="H", data=asteroids2);

plt.xlabel('Proximidade da Terra', fontsize=15)
plt.ylabel('Intensidade de Brilho dentro de 1UA', fontsize=15)
plt.title('Intensidade e proximidade do Asteroide', fontsize=16)
plt.show()
```

```python
# class = Classe do Asteroide com base a distancia e orbita
# H = Magnitude Absoluta, medida de brilho do objeto celeste 

plt.figure(figsize=(12,10))

sns.scatterplot(data=asteroids2, x="class", y="H", palette="deep", sizes=15)

plt.xlabel('Classe dos Asteroides', fontsize=15)
plt.ylabel('Intensidade de Brilho dentro de 1UA', fontsize=15)
plt.title('Relação de Classe e Brilho dos Asteroides', fontsize=16)
plt.show()
```
