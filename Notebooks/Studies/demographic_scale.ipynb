```python
import pandas as pd
import matplotlib.pyplot as plt
from typing import Dict
```

```python

data: Dict[str, list] = {'country': ['Belgian', 'India', 'United States', 'Brazil'],
                         'population': [11460000, 1353000000, 329634908, 209500000],
                         'color': ['red', 'green', 'blue', 'yellow']}
```

```python

for index, value in enumerate(data.get('country')):
    plt.bar(str(data.get('population')[index]),
             data.get('population')[index],
             color=data.get('color')[index],
             label=data.get('country')[index])

    plt.xlabel('Population')
    plt.ylabel('Scale per million')
    plt.title('Demographic Scale')
    plt.legend()
plt.show()
```
