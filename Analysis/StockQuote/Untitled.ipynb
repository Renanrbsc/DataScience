```python
# Imports
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
date_now = pd.to_datetime(dt.date.today()).strftime('%m/%d/%Y')
f'Current date: {date_now}'
```

```python
quote_bovespa = web.DataReader("^BVSP", data_source="yahoo", start="01/01/2020", end=date_now)
display(quote_bovespa)
```

```python
quote_bovespa["Adj Close"].plot(figsize=(15, 10))
plt.show()
```

```python
quote_petr = web.DataReader("PETR4.SA", data_source="yahoo", start="01/01/2020", end=date_now)
display(quote_petr)
```
